### Price Column/Columna de Precio:

*English:*

As noted above, the price column is currently displayed as text; this column includes the currency in words, which is why it's shown as text.

We will proceed as follows:
* Separating the currency and price into two columns
* Changing the price column to a number.

*Español:*

Como se observó en lineas anteriores, la columna precio está como texto; esta columna incluye la moneda con letras por lo que es debido a esto que se encuentra como texto.

Vamos a proceder de la siguiente manera:
* Separando la moneda y el precio en 2 columnas
* Cambiando la columna precio a número.

```python
#We check what currency formats are available before separating columns/Vemos qué formatos de moneda hay antes de separar
df_a["precio"].str.split(" ").str[0].unique()
```

```python
#Two new columns are created with values ​​from the price column/Se crean 2 columnas nuevas con valores de la columna precio
df_a["moneda"] = df_a["precio"].str.split(" ").str[0] #Currency column with the value before the space/Columna Moneda con el valor antes del espacio
df_a["precio_numero"] = df_a["precio"].str.split(" ").str[1] #precio_numero column with the value after the space/Columna precio_numero con el valor despues del espacio

df_a[["precio", "moneda", "precio_numero"]].head()
```

*English:*

We will now convert the precio_numero column to number format.

In the original df, the price is formatted with a period to separate thousands and a comma to separate decimals. We need to transform it into a format that Python recognizes better, which is: with no thousands separator and a period to separate decimals.

*Español:*

Procedemos ahora a convertir la columna precio_numero en formato número.

En el caso del df original el precio está escrito con formato de punto para separar miles y coma para separar decimales; debemos transformarlo a un formato que Python reconozca mejor, este es: sin nada para separar miles y punto para separar decimales.

```python
df_a["precio_numero"] = (
    df_a["precio_numero"]
    .str.replace(".", "", regex=False)   #Eliminate the period at thousands/Elimina el punto de los miles
    .str.replace(",", ".", regex=False)  #Converts the decimal comma to a period/Convierte la coma decimal en punto
    .astype(float) #Converts the value to a number/Transforma a número el valor
)

df_a[["precio", "precio_numero"]].head()
```

*English:*

The change in the way the data is written is observed; we will now evaluate if that column's format has changed:

*Español:*

Se observa el cambio de forma de escribir los datos, evaluaremos ahora si cambió de formato esa columna:

```python
df_a.info()
```

*English:*

The precio_numero column is shown in numeric format; we proceed to delete the price column to clean the data file.

*Español:*

Se observa la columna de precio_numero con formato numérico, procedemos a eliminar la columna precio para limpiar el df.

```python
df_a = df_a.drop(columns=["precio"])

df_a.head()
```

### Duplicates/Duplicados:

*English:*

Before finalizing the EDA, we will assess whether there are duplicates in the original database to remove them and simplify the analysis.

We will evaluate repetitions by title, author, and price; this way we ensure that we avoid, for example, the same book with different publishers and prices.

*Español:*

Antes de finalizar el EDA, evaluaremos si existen duplicados en el df original para eliminarlos y simplificar el análisis.

Evaluaremos repeticiones por título, autor y precio; de esta forma nos aseguramos que evitar por ejemplo, el mismo libro con distintas editoriales que tengan precios distintos.

```python
duplicados = df_a[df_a.duplicated(subset=["titulo", "autor", "precio_numero"], keep="first")]
duplicados.sort_values("titulo")
```

*English:*

In most cases, we see that the book is repeated within the same category, so it doesn't contribute to subsequent analyses.

We will evaluate the percentage of duplicates within the total data set:

*Español:*

En la mayoría de los casos vemos que el libro se repite dentro de la misma categoría por lo que no suma a análisis posteriores.

Evaluaremos el % de duplicados dentro del total de datos del df:

```python
total = len(df_a)
porcentaje = ((len(duplicados)) / total) * 100

print(f"Duplicados: {len(duplicados)}")
print(f"Total de filas: {total}")
print(f"Porcentaje de duplicados: {porcentaje:.2f}%")
```

*English:*

The decision to remove them is made for two reasons:
* The percentage of duplicates is small (less than 5%)
* Keeping these duplicates doesn't contribute to my analysis because they repeat data that clutters subsequent analyses.

We proceed to remove them:

*Español:*

Se decide eliminar por 2 motivos:
* El % de duplicados es pequeño (menor al 5%)
* Mantener estos duplicados no suman en mi análisis ya que repiten datos que ensucian los análisis posteriores.

Procedemos a eliminarlos:

```python
df_a = df_a.drop_duplicates(subset=["titulo", "autor", "precio_numero"], keep="first").copy()
df_a.shape
```

We re-evaluate the duplicates:

### Evaluation of Formats and Writing at Author Column:/Evaluación de Formatos y escritura en Columna Autor:

*English:*

The format in which the Authors are written is evaluated, and whether there is any extra space between the first and last name.

*Español:*

Se procede a evaluar el formato en que están escritos los Autores y si existiera algún espacio de más entre nombre y apellido.

```python
print(f"Autores únicos: {df_a['autor'].nunique()}")
df_a["autor"].unique()[:30]  #First 30, to avoid overloading the screen/Primeros 30, para no saturar la pantalla
```

*English:*

Two problems were observed:
1. Authors with one or more spaces between their first and last name, which can unnecessarily generate duplicate authors.

2. Formats written as Last Name, First Name and as First Name, Last Name.

The second problem is too difficult to solve, since attempting to resolve it automatically could produce errors in the results, providing incorrect information. Therefore, we will proceed to address the first error:

*Español:*

Se observan 2 problemas:
1. Autores con 1 o más espacios entre nombre y apellido, esto puede generar duplicados de autores de forma innecesaria.
2. Formatos escritos como Apellido, Nombre y como Nombre Apellido.

El segundo problema es demasiado dificil de solucionar, ya que si intentamos solucionarlo de forma automática podemos producir errores en los resultados de forma que nos entregue información incorrecta. Procedemos entonces a solucionar el 1er error:

```python
df_a["autor"] = df_a["autor"].str.replace(r"\s+", " ", regex=True).str.strip()
```

*English:*

We re-evaluated the list of unique authors to assess whether this change helped reduce potential duplicates:

*Español:*

Evaluamos de nuevo el listado de autores unicos para evaluar si este cambio ayudó a disminuir posibles duplicados:

```python
print(f"Autores únicos: {df_a['autor'].nunique()}")
df_a["autor"].unique()[:30]  ##First 30, to avoid overloading the screen/Primeros 30, para no saturar la pantalla
```

*English:*

It is observed that there were no changes regarding the number of authors, so the double spacing did not affect the previous count.

*Español:*

Se observa que no hubo cambios al respecto de la cantidad de autores, por lo que el doble espacio no afectaba el conteo anterior.
